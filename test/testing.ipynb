```python
import ziamath as zm
```

```python
# Alignment of fractions and enclosed fractions
s = '''
    <math>
      <mrow>
      <mo>(</mo>
        <mfrac linethickness="1">
          <mrow><mn>5</mn></mrow>
          <mrow><mn>2</mn></mrow>
        </mfrac>
      <mo>)</mo>
      </mrow>
      <mo>=</mo><mn>10</mn><mo>=</mo>
        <mfrac linethickness="1">
          <mrow><mn>5</mn></mrow>
          <mrow><mn>2</mn></mrow>
        </mfrac>
    </math>
'''
zm.Math(s)
```

```python
# Fenced operator with nothing in it
s = '''<math display='inline' xmlns='http://www.w3.org/1998/Math/MathML'>
   <msup>
      <mrow>
        <mo>|</mo>
      </mrow>
      <mrow>
         <mn>2</mn>
      </mrow>
   </msup>
   </math>'''
zm.Math(s)
```

```python
# Fenced
zm.Math('''<math>
  <mover>
    <mi>v</mi>
    <mo>&rarr;</mo>
  </mover>
  <mo>=</mo>
  <mfenced>
    <msub>
      <mi>v</mi>
      <mn>1</mn>
    </msub>
    <msub>
      <mi>v</mi>
      <mn>2</mn>
    </msub>
    <msub>
      <mi>v</mi>
      <mn>3</mn>
    </msub>
  </mfenced>
</math>''')
```

```python
zm.Math('''
<mfenced separators=',;:,'>
    <mo>a</mo>
    <mo>b</mo>
    <mo>c</mo>
    <mo>d</mo>
</mfenced>
''')
```

```python
# Comments in the XML, classnames in the tags
s = '''<math display='inline' xmlns='http://www.w3.org/1998/Math/MathML'>
<mi>e</mi>
<mo class='MathClass-rel'>=</mo>
<msub>
    <mrow>
        <mi>E</mi>
    </mrow>
    <mrow>
        <mi>m</mi>
    </mrow>
</msub>
<mi class='qopname'> cos</mi>
<mo> ⁡<!-- FUNCTION APPLICATION --> </mo>
<!-- nolimits -->
<mi mathvariant='italic'>ωt</mi>
</math>
'''
zm.Math(s)
```

```python
# Division sign
s = '''
<math xmlns="http://www.w3.org/1998/Math/MathML">
    <mrow>
        <mn>0</mn>
        <mo>÷</mo>
        <mn>23</mn>
    </mrow>
    <mo>=</mo>
    <mn>0</mn>
</math>'''
zm.Math(s)
```

```python
s = '''<math xmlns="http://www.w3.org/1998/Math/MathML">
            <mrow>
                <mi>x</mi>
                <mo>=</mo>
                <mfrac>
                    <mrow>
                        <mo form="prefix">-</mo>
                        <mi>b</mi>
                        <mo>±<!-- &amp;PlusMinus; --></mo>
                        <msqrt>
                            <msup>
                                <mi>b</mi>
                                <mn>2</mn>
                            </msup>
                            <mo>-</mo>
                            <mn>4</mn>
                            <mi>a</mi>
                            <mi>c</mi>
                        </msqrt>
                    </mrow>
                    <mrow>
                        <mn>2</mn>
                        <mi>a</mi>
                    </mrow>
                </mfrac>
            </mrow>
        </math>'''
zm.Math(s)
```

```python
import xml.etree.ElementTree as ET
from IPython.display import SVG
```

```python
# Rotations - default mode
# Blue dot (origin) should be on red square (ortho bounding box)
# (See https://matplotlib.org/stable/gallery/text_labels_and_annotations/demo_text_rotation_mode.html)
svg = ET.Element('svg')
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['xmlns:xlink'] = 'http://www.w3.org/1999/xlink'
svg.attrib['width'] = '500'
svg.attrib['height'] = '300'
svg.attrib['viewBox'] = '0 0 500 300'

for xy, halign, valign in [((20, 20), 'left', 'top'),
                           ((100, 20), 'center', 'top'),
                           ((180, 20), 'right', 'top'),
                           ((20, 80), 'left', 'center'),
                           ((100, 80), 'center', 'center'),
                           ((180, 80), 'right', 'center'),
                           ((20, 130), 'left', 'base'),
                           ((100, 120), 'center', 'base'),
                           ((180, 110), 'right', 'base'),
                           ((20, 180), 'left', 'bottom'),
                           ((100, 180), 'center', 'bottom'),
                           ((180, 180), 'right', 'bottom'),

                          ]:
    circ = ET.SubElement(svg, 'circle')
    circ.set('cx', str(xy[0]))
    circ.set('cy', str(xy[1]))
    circ.set('r', '3')
    circ.set('stroke', 'blue')
    circ.set('fill', 'blue')

    zm.Text('$\\sqrt{a^2}$\nABC', rotation=30,
                 size=12,
                 rotation_mode='default',
                 halign=halign, valign=valign).drawon(svg, xy[0], xy[1])

SVG(ET.tostring(svg))
```

```python
# Rotations - anchor mode
# Blue dot (origin) should be on blue square (tight/rotated bounding box)
# (See https://matplotlib.org/stable/gallery/text_labels_and_annotations/demo_text_rotation_mode.html)
svg = ET.Element('svg')
svg.attrib['xmlns'] = 'http://www.w3.org/2000/svg'
svg.attrib['xmlns:xlink'] = 'http://www.w3.org/1999/xlink'
svg.attrib['width'] = '500'
svg.attrib['height'] = '300'
svg.attrib['viewBox'] = '0 0 500 300'

for xy, halign, valign in [((20, 20), 'left', 'top'),
                           ((100, 20), 'center', 'top'),
                           ((180, 20), 'right', 'top'),
                           ((20, 80), 'left', 'center'),
                           ((100, 80), 'center', 'center'),
                           ((180, 80), 'right', 'center'),
                           ((20, 130), 'left', 'base'),
                           ((100, 120), 'center', 'base'),
                           ((180, 110), 'right', 'base'),
                           ((20, 180), 'left', 'bottom'),
                           ((100, 180), 'center', 'bottom'),
                           ((180, 180), 'right', 'bottom'),

                          ]:
    circ = ET.SubElement(svg, 'circle')
    circ.set('cx', str(xy[0]))
    circ.set('cy', str(xy[1]))
    circ.set('r', '3')
    circ.set('stroke', 'blue')
    circ.set('fill', 'blue')

    zm.Text('$\\sqrt{a^2}$\nABC', rotation=30,
                 size=12,
                 rotation_mode='anchor',
                 halign=halign, valign=valign).drawon(svg, xy[0], xy[1])

SVG(ET.tostring(svg))
```
